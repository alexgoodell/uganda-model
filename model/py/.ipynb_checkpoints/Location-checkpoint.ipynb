```python
# Import the data from the mohdhi2017 facilities list - using excel because CSV caused line-splitting problems (some cells have returns)
facilities = pd.read_excel("../../data/moh-div-health-info/facilities.xlsx", sheet="facilities")

# make splitter function
splitter = lambda x: pd.Series([i for i in reversed(x.split('/'))])
# split HSDT Code (see page 7 of mohdhi2017 for details of HSDT codes)
facilities[['parish_level_fac_id', 'parish_hsdt', 'subcounty_hsdt']] = facilities['HSDT Code'].apply(splitter)
facilities = facilities.drop(columns={'District_name'})

# Load the subcounty list
subcounties = pd.read_csv("../../data/moh-div-health-info/subcounties.csv")
rename = {'NHPI Code': 'subcounty_hsdt', 'Name': 'subcounty_name', 
          'HSDT Code': 'subcounty_hsdt_code', 'County': 'county_name', 
          'District': 'district_name',
          'Subregion': 'subregion_name'}
subcounties = subcounties.rename(columns=rename)
subcounties.to_csv("sc.csv")

# Add the subcounty data to the facilites list
p = pd.concat([facilities, subcounties], join='outer')
p
```
